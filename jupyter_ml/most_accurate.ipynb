```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import Imputer
from sklearn.model_selection import train_test_split
import numpy as np
import glob
```

```python
def datamanip(psg, hyp):
    
    csv_features=['EEGFpzCz', 'EEGPzOz', 'EOGhorizontal']
    csv_data=pd.read_csv(psg, names=csv_features)

    hyp_data=pd.read_csv(hyp, names=['EDFAnnotations'])
    hyp_data_manip=pd.DataFrame(columns=['EDFAnnotations'])

    X=csv_data[csv_features]
    y=hyp_data.EDFAnnotations

    big=len(csv_data.index)
    small=len(hyp_data.index)
    size=int(big/small)
    
    for i in range(0, len(y)):
        onePt=np.full(size, y[i])
        newDf=pd.DataFrame(onePt, columns=['EDFAnnotations'])
        hyp_data_manip = pd.concat([hyp_data_manip, newDf])

    if big > len(hyp_data_manip.index):
        X=X.iloc[(big-len(hyp_data_manip.index)):]

    y=hyp_data_manip.EDFAnnotations
    return X,y
```

```python
[X,y]=datamanip('psg/4002PSG.csv','Hypo/4002HYP.csv')
```

```python
name=['4011', '4012', '4021', '4022', '4031', '4032', '4041', '4042']
```

```python
for file in name:
    filenameH='Hypo/'+file+'HYP.csv'
    filenameP='psg/'+file+'PSG.csv'
    [C, D]=datamanip(filenameP, filenameH)
    X = pd.concat([X, C])
    y = pd.concat([y, D])
```

```python
my_imputer = Imputer()
X = my_imputer.fit_transform(X)

train_X, val_X, train_y, val_y=train_test_split(X, y, random_state=0)
```

```python
forest_model=RandomForestRegressor(random_state=1)
forest_model.fit(train_X, train_y)
```

```python
predicted_rate=forest_model.predict(val_X)
```

```python
print("the random forest error is ",mean_absolute_error(val_y, predicted_rate))
```
